```python
import yfinance as yf
from datetime import timedelta, date
import plotly.graph_objects as go

# Define today's date
today = date.today()

# Calculate start date (6 months ago)
start_date = today - timedelta(days=365 // 2)  # Integer division for full months

# Download AAPL data for the past 6 months
aapl_data = yf.download("AAPL", start=start_date.strftime('%Y-%m-%d'), end=today.strftime('%Y-%m-%d'))

# Create a Plotly figure
fig = go.Figure()

# Add line trace for closing price
fig.add_trace(go.Scatter(x=aapl_data.index, y=aapl_data['Close'], name='AAPL Closing Price'))

# Update layout with title and axis labels
fig.update_layout(
    title=f"Apple (AAPL) Closing Price - Past 6 Months",
    xaxis_title='Date',
    yaxis_title='Closing Price (USD)',
)

# Display the plot
fig.show()
```

```python
import yfinance as yf
from datetime import timedelta, date
import plotly.graph_objects as go

name = 'Emcor'
ticker = 'EME'


# Define today's date
today = date.today()
# Calculate start date (6 months ago)
start_date = today - timedelta(days=365 * 5)  # Integer division for full months
# Download AAPL data for the past 6 months
data = yf.download(ticker, start=start_date.strftime('%Y-%m-%d'), end=today.strftime('%Y-%m-%d'))



# Create a Plotly figure
fig = go.Figure()

# Add candlestick trace
fig.add_trace(go.Candlestick(
    x=data.index,
    open=data['Open'],
    high=data['High'],
    low=data['Low'],
    close=data['Close']
))

# Update layout with title and axis labels
fig.update_layout(
    title=str(name) + f" Candlestick Chart since " +str(start_date),
    xaxis_title='Date',
    yaxis_title='Price (USD)'
)

# Display the plot
fig.show()
```
